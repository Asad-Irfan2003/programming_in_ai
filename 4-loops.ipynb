# Loops in Python
- Loops execute a block of code repeatedly
- Python has two types of loops:
    1. For Loop
    2. While Loop
- All loops provide smilar basic functionality 

1. __For Loop:__
    - it is used to iterate over a sequence
    - Sequence can be a list, tuple, string, or range

```python
#Syntax:
#for variable in sequence:
    #Code to be executed for each element in sequence
```

```python
#Example: Printing each fruit in fruits list
fruits= ["apple","banana","cherry"]
for fruit in fruits:
    print(fruit)
```

__range() function:__
- It returns a sequence of numbers in a given range
- It is commonly used in for loops in python
- __Syntax:__
    - range(start, stop, step)
        - start(optional): Start value of sequence(0 by default)
        - stop: value before which sequence ends
        - step(optional): difference between each number in sequence(1 by default)
        - return value: object of sequence of numbers
- range(stop) is called when range is called with 1 argument
- range(start, stop) is called when range is called with 2 arguments
- range(start, stop, step) is called when range is called with 3 arguments

```python
#Example 1: printing number from 0 to 4
for i in range(5):
    print(i,end=" ")
```

```python
#Example 2: Printing Natural Numbers b/w
# 5 to 20(20 is exclusive)
for i in range(0,20):
    print(i,end=" ")
```

```python
#Example 3: Printing all even numbers b/w 0 and 9
for i in range(0,9,2):
    print(i,end=" ")
```

```python
#Example 4: Printing 5 to 20 in reverse
for i in range(20,5,-1):
    print(i, end=" ")
```

__NOTE:__
- range function only works with integers
- All arguments must be integers
- All arguments can be positive or negative
- step value cannot be zero otherwise it will give value error exception

2. __While Loop:__
    - It executes a sequence  of statements until a condition is satisfied


```python
#Syntax:
#while expression:
    #statement(s)
```

```python
#Example 1: printing 0 to 3
count=0
while(count<4):
    print(count)
    count=count+1
```

```python
#Example 2: Infinite while loop
count=0
while(count==0):
    print("Asad",end="")
```

### __Control Flow Statements__
- They alter the flow of loops
- 3 types
    1. break
        * Exits the loop permenantly
    2. continue
        * Skips the current iteration and moves to next one
    3. else
        * Optional clause when loop finishes normally (Without break)


```python
#Example of break
for i in range(10):
    if i==5:
        break
    print(i,end=" ")
```

```python
# Example of continue:
for i in range(10):
    if(i%2==0):
        continue
    print(i,end=" ")
```

```python
#Example of else:
for i in range(5):
    print(i,end=" ")
else:
    print("Loop completed",end=" ")
```
